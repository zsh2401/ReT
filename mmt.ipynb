```python
from pathlib import Path

mid_files = list(Path("dataset/nesmdb").glob("**/*.mid"))
len(mid_files)
```

```python
def time_to_beat(midi_data, time):
    """
    将时间 (秒) 转换为节拍位置
    Args:
        midi_data: PrettyMIDI 对象
        time: 当前时间 (秒)
    Returns:
        节拍位置
    """
    # 获取节奏变化（以微秒/拍为单位）
    tempo_changes, tempi = midi_data.get_tempo_changes()

    # 获取当前时间段对应的 tempo
    for i in range(len(tempo_changes) - 1):
        if tempo_changes[i] <= time < tempo_changes[i + 1]:
            tempo = tempi[i]
            break
    else:
        tempo = tempi[-1]  # 如果在最后一个区间

    # 获取节拍分母 (默认 4/4)
    time_signatures = midi_data.time_signature_changes
    time_signature = time_signatures[0] if time_signatures else None
    time_signature_denominator = time_signature.denominator if time_signature else 4

    # 换算节拍位置
    beats = time * (60 / tempo) * time_signature_denominator
    return beats
```

```python
import pretty_midi

# X_note = []
X_type = []
X_beat = []
X_position = []
X_pitch = []
X_duration = []
X_instrument = []
X = [X_type, X_beat, X_position, X_pitch, X_duration, X_instrument]
for file in mid_files[:1]:
    midi = pretty_midi.PrettyMIDI(str(file))
    # pretty_midi.
    for instrument in midi.instruments:
        for note in instrument.notes:
            X_type.append("note_on")
            X_beat.append(time_to_beat(midi, note.start))
            X_position.append(note.start )
            X_pitch.append(note.pitch)
            X_duration.append(note.end - note.start)
            instrument_name = pretty_midi.program_to_instrument_name(instrument.program)  # 乐器名
            X_instrument.append(instrument_name)
```
